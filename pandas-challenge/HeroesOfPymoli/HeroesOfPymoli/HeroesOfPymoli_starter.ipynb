```python
%matplotlib notebook
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
players_df = purchase_data.drop_duplicates("SN")
total_players = players_df["SN"].count()
print(f"Total players: {total_players}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
total_revenue = sum(purchase_data["Price"])
purchase_count = purchase_data["Purchase ID"].count()
average_price = total_revenue / purchase_count

# Create summary data frame to hold the results 
summary_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                            "Average Price": [average_price],
                           "Number of Purchases": [purchase_count],
                           "Total Revenue": [total_revenue]
})

# Format revenue and price with $ and two digits
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Count number of players by gender and rename the columns
gender_count = players_df["Gender"].value_counts().reset_index()
gender_count
gender_count.columns=["Gender","Total Count"]

# Calculate the percentage of players for each gender
gender_count["Percentage of Players"] = gender_count["Total Count"] / total_players * 100

# Format the percentage with % and two digits
gender_count["Percentage of Players"] = gender_count["Percentage of Players"].map("{:,.2f}%".format)
gender_count = gender_count.set_index("Gender")
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Group purchase data by gender
gender_df = purchase_data.groupby("Gender")

# Purchase count
gender_purchase = gender_df["Purchase ID"].count()
gender_purchase = pd.DataFrame(gender_purchase)
gender_purchase.columns = ["Purchase Count"]

# Calculate purchase total, avg. purchase price, avg. purchase total per person etc. by gender
gender_purchase["Purchase Total"] = gender_df["Price"].sum()
gender_purchase["Average Purchase Price"] = gender_purchase["Purchase Total"] / gender_purchase["Purchase Count"]
gender_purchase["Average Purchase Per Person"] = gender_purchase["Purchase Total"] / gender_count["Total Count"]

# Format price with $ and two digits
gender_purchase["Average Purchase Price"] = gender_purchase["Average Purchase Price"].map("${:,.2f}".format)
gender_purchase["Purchase Total"] = gender_purchase["Purchase Total"].map("${:,.2f}".format)
gender_purchase["Average Purchase Per Person"] = gender_purchase["Average Purchase Per Person"].map("${:,.2f}".format)
gender_purchase
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages using pd.cut()
min_age = purchase_data["Age"].min()
max_age = purchase_data["Age"].max()
age_bins = pd.cut(purchase_data["Age"], bins=[min_age-1,9,14,19,24,29,34,39,max_age], 
                  labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])

# Add column for age bins to the original data frames
purchase_data["Age Bins"] = age_bins
players_df=purchase_data.drop_duplicates("SN")

# Create a new data frame where data is grouped by age bins
age_df = players_df.groupby("Age Bins")

# Count number of players by age group
age_count = age_df["SN"].count()
age_count = pd.DataFrame(age_count)
age_count.columns = ["Total Count"]

# Calculate percentage of players in each age group
age_count["Percentage of Players"] = age_count["Total Count"] / total_players * 100
age_count

# Format the percentage with % and two digits
age_count["Percentage of Players"] = age_count["Percentage of Players"].map("{:,.2f}%".format)

# Display age demographic table
age_count
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchase_df = purchase_data.groupby("Age Bins")

# Purchase count
age_purchase = age_purchase_df["Purchase ID"].count()
age_purchase = pd.DataFrame(age_purchase)
age_purchase.columns = ["Purchase Count"]

# Calculate purchase total, avg. purchase price, avg. purchase total per person etc. by age
age_purchase["Purchase Total"] = age_purchase_df["Price"].sum()
age_purchase["Average Purchase Price"] = age_purchase["Purchase Total"] / age_purchase["Purchase Count"]
age_purchase["Average Purchase Per Person"] = age_purchase["Purchase Total"] / age_count["Total Count"]

# Format price with $ and two digits
age_purchase["Average Purchase Price"] = age_purchase["Average Purchase Price"].map("${:,.2f}".format)
age_purchase["Purchase Total"] = age_purchase["Purchase Total"].map("${:,.2f}".format)
age_purchase["Average Purchase Per Person"] = age_purchase["Average Purchase Per Person"].map("${:,.2f}".format)
age_purchase
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_df = purchase_data.groupby("SN")

# Purchase count
spenders_purchase = spenders_df["Purchase ID"].count()
spenders_purchase = pd.DataFrame(spenders_purchase)
spenders_purchase.columns = ["Purchase Count"]

# Calculate purchase total, avg. purchase price, avg. purchase total per person etc. by spenders
spenders_purchase["Purchase Total"] = spenders_df["Price"].sum()
spenders_purchase["Average Purchase Price"] = spenders_purchase["Purchase Total"] / spenders_purchase["Purchase Count"]
spenders_purchase = spenders_purchase.sort_values(by="Purchase Total", ascending=False).round(2)

# Format price with $ and two digits
spenders_purchase["Average Purchase Price"] = spenders_purchase["Average Purchase Price"].map("${:,.2f}".format)
spenders_purchase["Purchase Total"] = spenders_purchase["Purchase Total"].map("${:,.2f}".format)


spenders_purchase
```
